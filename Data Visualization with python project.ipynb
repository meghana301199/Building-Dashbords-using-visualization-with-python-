# Task 1.1 - Develop a Line plot using the functionality of pandas to show how automobile sales fluctuate from year to year.

Submit the image Line_plot_1.png

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generating sample data
np.random.seed(0)
years = range(2015, 2024)
sales = np.random.randint(500, 1500, size=len(years))

# Creating a DataFrame
df = pd.DataFrame({'Year': years, 'Sales': sales})

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Sales'], marker='o', linestyle='-')
plt.title('Automobile Sales Fluctuation (2015-2023)')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(df['Year'])
plt.tight_layout()
plt.show()
```

# Task 1.2 - Plot different lines for categories of vehicle type and analyse the trend to answer the question "Is there a noticeable difference in sales trends between different vehicle types during recession periods?"

Submit the image Line_plot_2.png

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generating sample data
np.random.seed(0)
years = range(2015, 2024)
types = ['Sedan', 'SUV', 'Truck']
sales_data = []

for vehicle_type in types:
    sales = np.random.randint(200, 800, size=len(years))
    sales_data.append(sales)

# Adding recession periods
recession_periods = [(2016, 2017), (2020, 2021)]

# Creating DataFrame
df = pd.DataFrame({'Year': np.repeat(years, len(types)),
                   'Vehicle Type': np.tile(types, len(years)),
                   'Sales': np.concatenate(sales_data)})

# Plotting
plt.figure(figsize=(10, 6))

for vehicle_type in types:
    data = df[df['Vehicle Type'] == vehicle_type]
    plt.plot(data['Year'], data['Sales'], marker='o', linestyle='-', label=vehicle_type)

# Highlighting recession periods
for period in recession_periods:
    plt.axvspan(period[0], period[1], color='red', alpha=0.2)

plt.title('Automobile Sales Fluctuation by Vehicle Type')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(df['Year'])
plt.legend()
plt.tight_layout()
plt.show()

```

# Task 1.3 - Use the functionality of Seaborn Library to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.

Submit the image Bar_Chart.png

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generating sample data
np.random.seed(0)
years = range(2015, 2024)
types = ['Sedan', 'SUV', 'Truck']
sales_data = []

for vehicle_type in types:
    sales = np.random.randint(200, 800, size=len(years))
    sales_data.append(sales)

# Adding recession periods
recession_periods = [(2016, 2017), (2020, 2021)]

# Creating DataFrame
df = pd.DataFrame({'Year': np.repeat(years, len(types)),
                   'Vehicle Type': np.tile(types, len(years)),
                   'Sales': np.concatenate(sales_data)})

# Adding recession flag
df['Recession'] = df['Year'].apply(lambda x: any(start <= x <= end for start, end in recession_periods))

# Creating the plot
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Vehicle Type', y='Sales', hue='Recession', ci=None)
plt.title('Comparison of Sales Trend per Vehicle Type for Recession and Non-Recession Periods')
plt.xlabel('Vehicle Type')
plt.ylabel('Sales')
plt.legend(title='Recession', labels=['Non-Recession', 'Recession'])
plt.tight_layout()
plt.savefig('Bar_Chart.png')
plt.show()

```

# Task 1.4 - Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.

Submit the imageSubplot.png

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample GDP data for illustration
years = range(2015, 2024)
gdp_non_recession = np.random.randint(1000, 2000, len(years))
gdp_recession = np.random.randint(800, 1500, len(years))

# Adding recession periods
recession_periods = [(2016, 2017), (2020, 2021)]

# Creating DataFrame
df_gdp = pd.DataFrame({'Year': years,
                       'GDP_Non_Recession': gdp_non_recession,
                       'GDP_Recession': gdp_recession})

# Adding recession flag
df_gdp['Recession'] = df_gdp['Year'].apply(lambda x: any(start <= x <= end for start, end in recession_periods))

# Plotting
fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Plot for non-recession period
df_non_recession = df_gdp[~df_gdp['Recession']]
axes[0].plot(df_non_recession['Year'], df_non_recession['GDP_Non_Recession'], marker='o', linestyle='-')
axes[0].set_title('GDP Variation during Non-Recession Period')
axes[0].set_ylabel('GDP')

# Plot for recession period
df_recession = df_gdp[df_gdp['Recession']]
axes[1].plot(df_recession['Year'], df_recession['GDP_Recession'], marker='o', linestyle='-')
axes[1].set_title('GDP Variation during Recession Period')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('GDP')

plt.tight_layout()
plt.show()

```

# Task 1.5 - Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.

Submit the image Bubble.png



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generating sample data
np.random.seed(0)
months = range(1, 13)  # Months of the year
seasonality = np.sin(np.pi * np.array(months) / 6)  # Simulating seasonal impact
sales = np.random.randint(500, 1500, size=len(months))

# Creating DataFrame
df_sales = pd.DataFrame({'Month': months, 'Sales': sales, 'Seasonality': seasonality})

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df_sales['Month'], df_sales['Sales'], s=df_sales['Sales'] / 50, c=df_sales['Seasonality'], cmap='coolwarm', alpha=0.7)
plt.colorbar(label='Seasonality')
plt.title('Impact of Seasonality on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(months)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Task 1.9 -  Develop a line plot to analyse the effect of the unemployment rate on vehicle type and sales during the Recession Period. 

Submit the image Line_plot_3.png

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data for illustration
# You should replace this with your actual data
months = range(1, 13)
unemployment_rate = np.random.uniform(5, 10, size=len(months))  # Sample unemployment rate data
sales = np.random.randint(100, 500, size=len(months))  # Sample sales data
vehicle_type = ['Sedan', 'SUV', 'Truck']  # Sample vehicle types

# Creating DataFrame
df = pd.DataFrame({'Month': months,
                   'Unemployment Rate': unemployment_rate,
                   'Sales': sales})

# Plotting
plt.figure(figsize=(10, 6))

# Plotting sales for each vehicle type
for vt in vehicle_type:
    df_vt = df.copy()
    df_vt['Vehicle Type'] = vt
    plt.plot(df_vt['Month'], df_vt['Sales'], label=vt)

# Plotting unemployment rate
plt.plot(df['Month'], df['Unemployment Rate'], color='black', linestyle='--', label='Unemployment Rate')

plt.title('Effect of Unemployment Rate on Vehicle Type and Sales During Recession Period')
plt.xlabel('Month')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

# Task 2.5 - Create and display graphs for Recession Report Statistics.

Submit the image RecessionReportgraphs.png

```python
import dash
from dash.dependencies import Input, Output
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go

# Sample options for vehicle type and year
vehicle_type_options = [
    {'label': 'Sedan', 'value': 'Sedan'},
    {'label': 'SUV', 'value': 'SUV'},
    {'label': 'Truck', 'value': 'Truck'}
]

year_options = [
    {'label': str(year), 'value': year} for year in range(2015, 2025)
]

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Recession Report Statistics"),  # Set the title of the dashboard
    
    # Dropdown for selecting vehicle type
    html.Label('Select Vehicle Type:'),
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=vehicle_type_options,
        value='Sedan'
    ),
    
    # Dropdown for selecting year
    html.Label('Select Year:'),
    dcc.Dropdown(
        id='year-dropdown',
        options=year_options,
        value=2015
    ),
    
    # Division for output display
    html.Div(id='output-display', className='output-container'),
    
    # Graph to display statistics
    dcc.Graph(id='recession-statistics-graph'),
])

# Define callback function
@app.callback(
    Output('recession-statistics-graph', 'figure'),
    [Input('vehicle-type-dropdown', 'value'),
     Input('year-dropdown', 'value')]
)
def update_graph(vehicle_type, year):
    # Replace this with your actual data processing and graph creation logic
    x_data = [1, 2, 3, 4, 5]
    y_data = [10, 20, 30, 40, 50]
    
    data = [
        go.Bar(
            x=x_data,
            y=y_data,
            marker=dict(color='blue')
        )
    ]
    
    layout = go.Layout(
        title=f'Statistics for {vehicle_type} in {year}',
        xaxis=dict(title='X Axis Title'),
        yaxis=dict(title='Y Axis Title'),
        barmode='group'
    )
    
    fig = go.Figure(data=data, layout=layout)
    
    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
